Using Weighted Average for each movies's average rating
W=(Rv+Cm)/(v+m)
where, W=Weighted Rating
       R=average for the movie as a number of 0 to 10(mean)= Rating
       v=number of votes for the movie=(votes)
       m=minimum number of votes required to be listed in the top 250(currently 3000)
       C=the mean vote across the whole report(currently 6.9)
       
Source: http://trailerpark.weebly.com/imdb-rating.html?source=post_page

```python
# Calculate all the components based on the above formula
v=movies_rating['vote_count']
R=movies_rating['vote_average']
C=movies_rating['vote_average'].mean()
m=movies_rating['vote_count'].quantile(0.70)
```

```python
movies_rating['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
movies_rating.head()
```

```python
movie_sorted_ranking=movies_rating.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['title', 'vote_count', 'vote_average', 'weighted_average', 'popularity','genres']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['title'].head(10), data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_rating[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_rating[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movies_rating.head()
```

```python
movies_rating['score'] = movies_rating['normalized_weight_average'] * 0.5 + movies_rating['normalized_popularity'] * 0.5
movies_scored_df = movies_rating.sort_values(['score'], ascending=False)
popular_movies=movies_scored_df[['movie_id','title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
movies_scored_df['genres']=movies_scored_df['genres'].apply(lambda x:" ".join(x)) #converting the list into a string
movies_scored_df['genres']=movies_scored_df['genres'].apply(lambda x:x.lower())
movies_scored_df['cast']=movies_scored_df['cast'].apply(lambda x:" ".join(x)) #converting the list into a string
movies_scored_df['cast']=movies_scored_df['cast'].apply(lambda x:x.lower())
```
